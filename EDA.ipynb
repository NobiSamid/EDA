```python
pln=pd.read_csv(r"E:\New folder\eda\planes.csv")
pln
```

```python
# Find the five percent threshold
threshold = len(pln) * 0.05

# Count the number of missing values in each column
print(pln.isna().sum(), threshold)
```

```python
import re
```

```python
len(pln)*.05
```

```python
# Create a filter
cols_to_drop = pln.columns[pln.isna().sum() <= threshold]

# Drop missing values for columns below the threshold
pln.dropna(subset=cols_to_drop, inplace=True)

print(pln.isna().sum())
```

```python
# Calculate median plane ticket prices by Airline
airline_prices = pln.groupby("Airline")["Price"].median()

print(airline_prices)
```

```python
#Convert to a dictionary
prices_dict = airline_prices.to_dict()
prices_dict
```

```python

# Map the dictionary to missing values of Price by Airline
pln['Price']= pln["Price"].fillna(pln["Airline"].map(prices_dict)) 
pln.isna().sum()
```

```python
# pln.drop("Additional_Info", axis=1, inplace=True)
pln.isna().sum()
```

```python
# Filter the DataFrame for object columns
non_numeric = pln.select_dtypes("object")
print(non_numeric)
# Loop through columns
for i in non_numeric.columns:
  
  # Print the number of unique values
  print(f"Number of unique values in {i} column: ", non_numeric[i].nunique())
```

```python
# Create a list of categories
flight_categories = ["Short-haul", "Medium", "Long-haul"]

# Create short_flights
short_flights = "^0h|^1h|^2h|^3h|^4h"

# Create medium_flights
medium_flights = "^5h|^6h|^7h|^8h|^9h"

# Create long_flights
long_flights = "10h|11h|12h|13h|14h|15h|16h"
```

```python
# Create conditions for values in flight_categories to be created
conditions = [
    (pln["Duration"].str.contains(short_flights)),
    (pln["Duration"].str.contains(medium_flights)),
    (pln["Duration"].str.contains(long_flights))
]

# Apply the conditions list to the flight_categories
pln["Duration_Category"] = np.select(conditions, 
                                        flight_categories,
                                        default="Extreme duration")

# Plot the counts of each category
sns.countplot(data=pln, x="Duration_Category")
plt.show()
```

```python
# Preview the column
print(pln["Duration"].head())

# Remove the string character
pln["Duration"] = pln["Duration"].str.replace(" ", "")

# Convert to float data type
# pln["Duration"] = pln["Duration"].astype(float)

# Plot a histogram
# sns.histplot(data=pln, x='Duration')
# plt.show()
print(pln["Duration"].head())
```

```python
def duration_to_hours(duration):
    if pd.isna(duration):  # Skip NaN
        return np.nan
    duration = str(duration).strip()  # Convert to string
    match = re.match(r'(?:(\d+)h)?\s*(?:(\d+)m)?', duration)
    if not match:
        return np.nan
    hours = int(match.group(1)) if match.group(1) else 0
    minutes = int(match.group(2)) if match.group(2) else 0
    return hours + minutes / 60

pln["duration_hours"] = pln["Duration"].apply(duration_to_hours).astype(float)

pln
```

```python
sns.histplot(data=pln, x='duration_hours')
plt.show()
```

```python
# Price standard deviation by Airline
pln["airline_price_st_dev"] = pln.groupby("Airline")["Price"].transform(lambda x: x.std())

pln[["Airline", "airline_price_st_dev", "Price"]].value_counts()
```

```python
# Find the 75th and 25th percentiles
price_seventy_fifth = pln["Price"].quantile(0.75)
price_twenty_fifth = pln["Price"].quantile(0.25)

# Calculate iqr
prices_iqr = price_seventy_fifth - price_twenty_fifth

# Calculate the thresholds
upper = price_seventy_fifth + (1.5 * prices_iqr)
lower = price_twenty_fifth - (1.5 * prices_iqr)

# Subset the data
pln = pln[(pln["Price"] > lower) & (pln["Price"] < upper)]

print(pln["Price"].describe())
```

```python
slr = pd.read_csv(r"E:\New folder\eda\ds_salaries_clean.csv")
slr
```

```python
# Cross-tabulate Company_Size and Experience
print(pd.crosstab(slr["Company_Size"], slr["Experience"]))
```

```python
# Cross-tabulate Job_Category and Company_Size
print(pd.crosstab(slr["Experience"], slr["Company_Size"],
            values=slr["Salary_USD"], aggfunc="mean"))
```

```python
# Get the month of the response
slr["month"] = slr["date_of_response"].dt.month

# Extract the weekday of the response
slr["weekday"] = slr['date_of_response'].dt.weekday

# Create a heatmap
sns.heatmap(slr.corr(numeric_only=True), annot=True)
plt.show()
```

```python
# Find the 25th percentile
twenty_fifth = slr["Salary_USD"].quantile(0.25)

# Save the median
salaries_median = slr["Salary_USD"].quantile(.5)

# Gather the 75th percentile
seventy_fifth = slr['Salary_USD'].quantile(.75)
print(twenty_fifth, salaries_median, seventy_fifth)
```

```python
# Create salary labels
salary_labels = ["entry", "mid", "senior", "exec"]

# Create the salary ranges list
salary_ranges = [0, twenty_fifth, salaries_median, seventy_fifth, slr["Salary_USD"].max()]

# Create salary_level
slr["salary_level"] = pd.cut(slr["Salary_USD"],
                                  bins=salary_ranges,
                                  labels=salary_labels)

# Plot the count of salary levels at companies of different sizes
sns.countplot(data=slr, x="Company_Size", hue="salary_level")
plt.show()
```
